```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hue_order = ['cm', 'ocm', 'cm2', 'ccm', 'occm', 'ccm2']
```

```python
df = pd.read_csv('result1.csv')
df.info()
```

```python
df.fasta_file = df.fasta_file.str.replace('input/', '')
df.fasta_file = df.fasta_file.str.replace('.fa', '')
```

```python
fig, ax = plt.subplots(1,1,figsize=(12, 5))
sns.barplot(x='fasta_file', y='rms_error', hue='method', data=df, ax = ax, palette='tab10',hue_order= hue_order)
ax.set_title('RMS error of different methods')
```

```python
fig, ax = plt.subplots(1,1,figsize=(12, 5))
sns.barplot(x='fasta_file', y='corr_coef', hue='method', data=df, ax = ax, palette='tab10',hue_order= hue_order)
ax.set_title('Correlation Coefficient for different methods')
plt.ylim(0.6, 0.95)
```
